# Checklist Benchmark Case

This notebook is the executed companion artifact for the Composition module. It uses an anonymized financial product-disclosure checklist benchmark.

```python
card = {
    "assess": [
        "Long-form source document",
        "Schema-constrained checklist output",
        "Coverage and traceability risk",
        "Human review required before production use",
    ],
    "route": [
        "Reasoning x Chain for baseline",
        "Reflection x Chain / Loop for repair",
        "Governance x Route for candidate review",
        "Parallel -> Route for high-recall queue",
    ],
    "select": [
        "single_pass",
        "critique_repair",
        "iterative_self_refine",
        "candidate_guided_review",
        "coverage_preserving_union_queue",
        "orchestrated_consensus_refine",
    ],
}
for step, values in card.items():
    print(f"{step.upper()}: " + " | ".join(values))
```

```python
rows = [
    {"pattern_id": "single_pass", "pattern_label": "Reasoning x Chain", "matched_over_gold": "6 / 12", "precision": 0.375, "recall": 0.5, "f1": 0.4286},
    {"pattern_id": "critique_repair", "pattern_label": "Reflection x Chain", "matched_over_gold": "7 / 12", "precision": 0.4375, "recall": 0.5833, "f1": 0.5},
    {"pattern_id": "iterative_self_refine", "pattern_label": "Reflection x Loop", "matched_over_gold": "7 / 12", "precision": 0.4375, "recall": 0.5833, "f1": 0.5},
    {"pattern_id": "candidate_guided_review", "pattern_label": "Governance x Route", "matched_over_gold": "9 / 12", "precision": 0.5625, "recall": 0.75, "f1": 0.6429},
    {"pattern_id": "coverage_preserving_union_queue", "pattern_label": "Composite: Parallel -> Route", "matched_over_gold": "10 / 12", "precision": 0.3448, "recall": 0.8333, "f1": 0.4878},
    {"pattern_id": "orchestrated_consensus_refine", "pattern_label": "Composite: Parallel -> Route -> Loop", "matched_over_gold": "8 / 12", "precision": 0.5, "recall": 0.6667, "f1": 0.5714},
]
columns = ["pattern_id", "pattern_label", "matched_over_gold", "precision", "recall", "f1"]
print("| " + " | ".join(columns) + " |")
print("| " + " | ".join(["---"] * len(columns)) + " |")
for row in rows:
    print("| " + " | ".join(str(row[col]) for col in columns) + " |")
```

```python
best = {"pattern_id": "coverage_preserving_union_queue", "matched_over_gold": "10 / 12"}
conclusion = "Preserve a high-recall review queue before final checklist compression; the queue exposes complementary coverage that a compact final answer can hide."
print(f"Best pattern: {best['pattern_id']} -> {best['matched_over_gold']}")
print(conclusion)
```
